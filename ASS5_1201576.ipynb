```python
import pandas as pd
import numpy as np
import random
from itertools import combinations
from collections import Counter
```

```python
#read data fream
df=pd.read_csv("online_retail_II.csv")
df.sort_values(by="Quantity")
```

Randomly select 200k row from the dataset using your ID as seed.

```python
# Select 200000 random rows
np.random.seed(1201576)
sample_200K=df.sample(n=200000,replace=False)
```

1.   Determine the total sales per invoice and identify the top 5 invoices.

```python
sample_200K['TotalSales'] = sample_200K['Quantity'] * sample_200K['Price']
```

```python
top_5_invoices = sample_200K.groupby('Invoice')['TotalSales'].sum().sort_values(ascending=False).head(5)
top_5_invoices_df=top_5_invoices.reset_index()
print(top_5_invoices_df)
```

2.Compute the total sales per customer and list the top 5 customers.


```python

top_5_customer = sample_200K.groupby("Customer ID")['TotalSales'].sum().sort_values(ascending=False).head(5)

top_5_customer_df = top_5_customer.reset_index()

print('Top 5 Customers by Total Sales:')
print(top_5_customer_df)
```

This output indicates the top 5 customers by total sales, with each line showing a customer ID followed by their corresponding total sales amount.

Identify the top 5 selling items during this period.


```python
top_5_items=sample_200K.groupby("StockCode")['Quantity'].sum().sort_values(ascending=False).head(5)
top_5_items_df = top_5_items.to_frame().reset_index()
print("Top 5 Items by Total Sales:")
print(top_5_items_df)
```

 output shows the top 5 items by total sales, where each row represents an item identified by its 'StockCode' and its corresponding total sales quantity ('Quantity').

Determine the most frequently purchased item pairs by customers and display the top 20 pairs.

```python
grouped = sample_200K.groupby('Invoice')['Description'].apply(list)
pairs = Counter()
for items in grouped:
    for pair in combinations(set(items), 2):
        pairs[pair] += 1

top_20_pairs = pairs.most_common(20)
top_20_pairs_df = pd.DataFrame(top_20_pairs, columns=['Pair', 'Count'])

top_20_pairs_df
```

 code calculates the top 20 pairs of items based on their occurrence in invoices. Each pair consists of two items

The retail store aims to introduce a new promotion targeting items with low sales volume. This promotion entails offering a discounted price for a pair of two different items, with one item being among the top 10 best sellers and the other from the bottom 10 sellers. Customers who purchase these two items together will qualify for the special pricing. Your task is to recommend 10 pairs of items to kickstart this campaign.

```python
top_10_items = df.groupby('Description')['Quantity'].sum().nlargest(10).index.tolist()
bottom_10_items = df.groupby('Description')['Quantity'].sum().nsmallest(10).index.tolist()
promotion_pairs = [(top, bottom) for top in top_10_items for bottom in bottom_10_items][:10]

top_10_items_df = pd.DataFrame(promotion_pairs, columns=['Description','Description2'])
top_10_items_df
```

the output show pairs of items from the top 10 and bottom 10 items, and to recommend 10 pairs of items to kickstart the campaign
