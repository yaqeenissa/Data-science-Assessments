```python
average_score_by_region = df.groupby('Region')['Happiness Score'].mean().sort_values()

highest_region = average_score_by_region.idxmax()
lowest_region = average_score_by_region.idxmin()

plt.figure(figsize=(10, 6))
average_score_by_region.plot(kind='bar', color='pink')
plt.title('Average Happiness Score by Region')
plt.xlabel('Region')
plt.ylabel('Average Happiness Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print(f"The region with the highest average happiness score is: {highest_region}")
print(f"The region with the lowest average happiness score is: {lowest_region}")
```

The region with the highest average happiness score is: Australia and New Zealand
The region with the lowest average happiness score is: Sub-Saharan Africa

```python
regions_of_interest = [
    "Australia and New Zealand",
    "Latin America and Caribbean",
    "Sub-Saharan Africa",
    "Southern Asia"
]

for region in regions_of_interest:

    region_df = df[df['Region'] == region]
    plt.figure(figsize=(7, 8))
    countries_in_region = region_df['Country'].unique()


    sns.swarmplot(x='Country', y='Happiness Score', data=region_df, palette='Set1', order=countries_in_region)
    plt.title(f' Happiness Score for {region}')
    plt.xlabel('Country')
    plt.ylabel('Happiness Score')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

```

From the visualization, we can observe the distribution of Happiness Scores within each region and identify countries with outlier scores that fall significantly above or below the typical range for that region.

* For example, in the region "Latin America and Caribbean," most countries have scores in the range of 4.9 to 7.2, but Haiti stands out as an outlier with a score of 3.5.




```python
average_scores_by_year = df.groupby('year')['Happiness Score'].mean()

plt.figure(figsize=(8, 6))
plt.plot(average_scores_by_year.index, average_scores_by_year.values, marker='o', linestyle='-')

plt.title('Global Happiness Score Over Time')
plt.xlabel('Year')
plt.ylabel('Average Happiness Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```
